# Loading Dataset

##### Importing libraries 

```python
import os
import pandas as pdb
```

##### Importing the dataset from subdirectories and extracting the activites names

```python
path = r"C:\Users\Administrator\Desktop\AS1OSMAN\trimmed_raw_data"

dfs = []

for (dirpath, dirnames, filenames) in os.walk(path):
    for filename in filenames:
        file = os.path.join(dirpath, filename)
        if file.endswith('.csv'):
            temp_df = pdb.read_csv(file)
            
            # Extracting the activity name from the directory path
            activity_name = os.path.basename(dirpath)
            
            temp_df['Activity'] = activity_name.split('.', 1)[-1]  # Remove the leading number and dot
            dfs.append(temp_df)
            
            print(f'File {file} loaded as CSV.')
```

```python
dfs
```

##### Setting column names

```python
for df in dfs:
    new_column_names = ['timelapseAcc', 'accelometerX', 'accelometerY', 'accelometerZ', "timelapseGyro", "gyroX", "gyroY", "gyroZ", 'Activity']
    # Select all columns except 'Activity' column
    columns_to_rename = [col for col in df.columns if col != 'Activity']
    # Rename columns excluding 'Activity'
    df.rename(columns=dict(zip(columns_to_rename, new_column_names[:-1])), inplace=True)
```
